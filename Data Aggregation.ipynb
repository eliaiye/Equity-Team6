```python
#Calculate return from price
df_rtn=priceTable.pct_change()
df_rtn
```

```python
Dates=Dates.astype('datetime64')
Dates
```

```python
#Create dataframes for each stock, with date as index, factors as column
Symb=df_rtn.columns
stocks={}
for s in Symb:
    indFactor=(Sect[str(s)]==Sect.unique())*1
    indFactor=np.array(list(indFactor)*len(Dates))
    stocks[str(s)]=pd.DataFrame(indFactor.reshape(len(Dates),-1),columns=list(Sect.unique()),index=Dates)  
```

```python
#Mean Reversion Signal
df_rtn_prev=df_rtn.shift(1)
df_5d_rolling=df_rtn_prev.rolling(5,win_type=None).sum()
df_5d_rolling.dropna(inplace=True)
```

```python
mr_sig_raw=-df_5d_rolling
mr_sig_mean=mr_sig_raw.mean(axis=1)
mr_sig_mean.head()
```

```python
mr_sig_std=mr_sig_raw.std(axis=1)
#normalize
mr_sig_norm = mr_sig_raw.sub(mr_sig_mean,axis=0).div(mr_sig_std,axis=0)
mr_sig_norm.tail(3)
```

```python
mom_sig_raw = df_rtn.shift(22).rolling(231).sum().dropna()
mom_sig_norm=mom_sig_raw.sub(mom_sig_raw.mean(axis=1),axis=0).div(mom_sig_raw.std(axis=1),axis=0)
mom_sig_norm.dropna()
```

```python
# Add to stock dataframes
for s in Symb:
    MR=mr_sig_norm[s]
    MOM=mom_sig_norm[s]
    Return=df_rtn[s]
    stocks[str(s)]["MR"]=MR 
    stocks[str(s)]["MOM"]=MOM
    stocks[str(s)]['Return']=Return
    stocks[str(s)]=stocks[str(s)].loc["2015-01-31":]
```

```python
#pb, pe facors
pb=pd.read_csv("pb_final.csv",index_col=0,parse_dates=True)
pe=pd.read_csv("pe_final.csv",index_col=0,parse_dates=True)
```

```python
#RoE factor
roe=pd.read_csv("roe_final.csv",index_col=0,parse_dates=True)
roe
```

```python
growth=pd.read_csv("new_pe.csv",index_col=0,usecols=['Unnamed: 0', 'growth factor'])
growth
```

```python
#Add all factors to stock dataframes
for i in range(408):
    testdf=pd.read_csv('stocks/'+str(Symb[i])+'.csv',usecols=['Period End Date','Net Margin','Total Asset turnover'])
    testdf['Period End Date']=testdf['Period End Date'].astype('datetime64')
    testdf=testdf.set_index('Period End Date')
    stocks[Symb[i]]=pd.concat([stocks[Symb[i]],testdf],axis=1)
    stocks[Symb[i]][["Net Margin",'Total Asset turnover']]=stocks[Symb[i]][["Net Margin",'Total Asset turnover']].fillna(method='ffill')
    
    roe_=roe[Symb[i]]
    roe_.name="Return on Equity"
    stocks[Symb[i]]=pd.concat([stocks[Symb[i]],roe_],axis=1)
    stocks[Symb[i]]["Return on Equity"]=stocks[Symb[i]]["Return on Equity"].fillna(method="ffill")
    
    DCF_=dcf_prem[Symb[i]]
    DCF_.name="DCF Prem"
    stocks[Symb[i]]=pd.concat([stocks[Symb[i]],DCF_],axis=1)
    stocks[Symb[i]]["DCF Prem"]=stocks[Symb[i]]["DCF Prem"].fillna(method="ffill")
    
    PE_=pe_prem[Symb[i]]
    PE_.name="PE Prem"
    stocks[Symb[i]]=pd.concat([stocks[Symb[i]],PE_],axis=1)
    stocks[Symb[i]]["PE Prem"]=stocks[Symb[i]]["PE Prem"].fillna(method="ffill")
    
    price_=price_df[Symb[i]]
    price_.name="Price"
    stocks[Symb[i]]=pd.concat([stocks[Symb[i]],price_],axis=1)
    stocks[Symb[i]]["Price"]=stocks[Symb[i]]["Price"].fillna(method="ffill")
    
    stocks[Symb[i]]["Growth"]=[growth.loc[Symb[i],"growth factor"]]*stocks[Symb[i]].shape[0]
      
    pbb=pb[Symb[i]]
    pbb.name="PB"
    pee=pe[Symb[i]]
    pee.name="PE"
    stocks[Symb[i]]=pd.concat([stocks[Symb[i]],pbb],axis=1)
    stocks[Symb[i]]["PB"]=stocks[Symb[i]]["PB"].fillna(method="ffill")
    stocks[Symb[i]]=pd.concat([stocks[Symb[i]],pee],axis=1)
    stocks[Symb[i]]["PE"]=stocks[Symb[i]]["PE"].fillna(method="ffill")
    stocks[Symb[i]].dropna(inplace=True)
```

```python
# Create factor dataframes
factors={}
for f_ in f:
    dff=pd.DataFrame()
    for s in Symb:
        series=stocks[s][f_]
        series.name=s
        dff=pd.concat([dff,series],axis=1)
    factors[f_]=dff
factors["MOM"]
```

```python
#Standardize all fundamental factors
factors['Net Margin']=factors['Net Margin'].sub(factors['Net Margin'].mean(axis=1),axis=0).div(factors['Net Margin'].std(axis=1),axis=0)
factors['Total Asset turnover']=factors['Total Asset turnover'].sub(factors['Total Asset turnover'].mean(axis=1),axis=0).div(factors['Total Asset turnover'].std(axis=1),axis=0)
factors['Return on Equity']=factors['Return on Equity'].sub(factors['Return on Equity'].mean(axis=1),axis=0).div(factors['Return on Equity'].std(axis=1),axis=0)
factors['PE Prem']=factors['PE Prem'].sub(factors['PE Prem'].mean(axis=1),axis=0).div(factors['PE Prem'].std(axis=1),axis=0)
factors['Growth']=factors['Growth'].sub(factors['Growth'].mean(axis=1),axis=0).div(factors['Growth'].std(axis=1),axis=0)
factors['Price']=factors['Price'].sub(factors['Price'].mean(axis=1),axis=0).div(factors['Price'].std(axis=1),axis=0)
factors['DCF Prem']=factors['DCF Prem'].sub(factors['DCF Prem'].mean(axis=1),axis=0).div(factors['DCF Prem'].std(axis=1),axis=0)
factors['PB']=factors['PB'].sub(factors['PB'].mean(axis=1),axis=0).div(factors['PB'].std(axis=1),axis=0)
factors['PE']=factors['PE'].sub(factors['PE'].mean(axis=1),axis=0).div(factors['PE'].std(axis=1),axis=0)
```

```python
def score(result):
    '''
    input: list of classifier results for each header
    output: score
    '''
    n=0
    p=0
    for i in range(len(result)):
        if result[i]["label"]=="NEGATIVE":
            n+=1
        else:
            p+=1
    return(p/(n+p))
score(result)
```
