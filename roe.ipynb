```python
import pandas
import re
from bs4 import BeautifulSoup as bs
import json
import requests
from requests.exceptions import RequestException
import time
import pandas as pd
```

```python
df = pd.read_csv("t440.csv",index_col=0)
```

```python
def get_one_page(url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.162 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            return response.text
        return None
    except RequestException:
        return None
```

```python
def parse_one_page(html,ticker):
    time = []
    factor = []
    items = re.findall('<td style="text-align:center;">(.*)</td>', html)
    for i in range(30):
        time.append(items[i*4])
        factor.append(items[i*4+3].strip("%"))
    dic = {
        "time":time,
        ticker:factor
    }
    return dic
```

```python
dictionary = {}
for i in range(len(df)):
    ticker = df["Symbol"][i]
    name = df["name"][i]
    url = "https://www.macrotrends.net/stocks/charts/{}/{}/roe".format(ticker,name)
    #print(url)
    try:
        html = get_one_page(url)
        
        d = parse_one_page(html,ticker)
        dictionary.update(d)
    except:
        print(ticker)
roe = pd.DataFrame(dictionary)    
roe.to_csv("roe_macro.csv") 
roe
```
