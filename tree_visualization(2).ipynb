```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
```

```python
import pickle


pkl_filename = "pickle_model.pkl"

# Load from file
with open(pkl_filename, 'rb') as file:
    pickle_model = pickle.load(file)
```

```python
import graphviz
clf = pickle_model
```

```python
feature_names = ['Industrials', 'Health Care', 'Information Technology',
       'Communication Services', 'Consumer Discretionary', 'Utilities',
       'Financials', 'Materials', 'Real Estate', 'Consumer Staples', 'Energy',
       'MR', 'MOM', 'Net Margin', 'Total Asset turnover', 'Return on Equity',
       'DCF Prem', 'PE Prem', 'Growth', 'Price', 'PE', 'PB']
```

```python
class_names = "Excess Return"
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=feature_names,  
                                class_names=class_names,
                                filled=True)

graph = graphviz.Source(dot_data, format="png") 
graph
```

```python
fig = plt.figure(figsize=(25,20))
annotations= tree.plot_tree(clf, 
                   feature_names=feature_names,  
                   class_names=class_names,
                   filled=True)
```
